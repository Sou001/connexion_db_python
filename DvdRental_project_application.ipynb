```python
%%sql
select * from actor
limit 5;
```

```python
%%sql
select first_name || " " || last_name as Actor_Name from actor
limit 5;
```

```python
%%sql
select lower(first_name) || "." || upper(last_name) as Actor_Name from actor
limit 5;
```

```python
%%sql
select upper(last_name) || "." || upper(substr(first_name,1,1)) || lower(substr(first_name,2,length(first_name))) as Actor_Name from actor
limit 5;
```

```python
%%sql
select * 
from actor
where upper(first_name) = "JENNIFER"
limit 5;
```

```python
%%sql
select * 
from actor
where first_name like '___'
```

```python
%%sql
select * 
from actor
where length(first_name) = 3
```

```python
%%sql
select actor_id, first_name, last_name, length(first_name) nb_car_first, 
length(last_name) nb_car_last
from actor
order by nb_car_first desc
```

```python
%%sql
select actor_id, first_name, last_name, length(first_name) nb_car_first, 
length(last_name) nb_car_last
from actor
order by nb_car_first desc, nb_car_last
```
